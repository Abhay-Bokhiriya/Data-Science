```python
# Frequency/count encoding for Mfr Name
encoder = ce.CountEncoder(cols=['Mfr Name'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Frequency/count encoding for Division
encoder = ce.CountEncoder(cols=['Division'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Ordinal encoding for air aspiration 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Air Aspiration Method Desc', 'mapping': {
                                    'Supercharged': 1, 'Turbocharged': 2, 
                                    'Other':3, 'Naturally Aspirated':4}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Frequency/count encoding for carline
encoder = ce.CountEncoder(cols=['Carline'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Ordinal encoding for Trans Desc 
new_data['Trans Desc'].unique()
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Trans Desc', 'mapping': {
                                    'Manual': 1, 'Automated Manual': 2, 
                                    'Semi-Automatic':3, 'Automated Manual- Selectable (e.g. Automated Manual with paddles)':4,
                                    'Automatic': 5,'Selectable Continuously Variable (e.g. CVT with paddles)': 6,
                                    'Continuously Variable':7 }}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Frequency/count encoding for Verify Mfr Cd
encoder = ce.CountEncoder(cols=['Verify Mfr Cd'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Lockup Torque Converter 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Lockup Torque Converter',
                                      'mapping': {'Y': 0, 'N': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Lockup Torque Converter 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Drive Sys', 
                                      'mapping': {'A': 0, 'R': 1, 'F': 2, '4': 3, 'P': 4}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Frequency/count encoding for Calc Approach Desc
encoder = ce.CountEncoder(cols=['Calc Approach Desc'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Trans Creeper Gear 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Trans Creeper Gear', 
                                      'mapping': {'Y': 0, 'N': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Fuel Usage Desc - Conventional Fuel
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Fuel Usage Desc - Conventional Fuel', 
                            'mapping':{'Gasoline (Regular Unleaded Recommended)': 0,
                                    'Gasoline (Premium Unleaded Required)': 1,
                                    'Gasoline (Premium Unleaded Recommended)': 2,
                                    'Diesel, ultra low sulfur (15 ppm, maximum)': 3,
                                    'Gasoline (Mid Grade Unleaded Recommended)': 4}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Gas Guzzler Exempt (Where Truck = 1975 NHTSA truck definition)
new_data['Gas Guzzler Exempt (Where Truck = 1975 NHTSA truck definition)'].unique()
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Gas Guzzler Exempt (Where Truck = 1975 NHTSA truck definition)', 
                                      'mapping': {'T': 0, 'N': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Frequency/count encoding for Carline Class Desc
encoder = ce.CountEncoder(cols=['Carline Class Desc'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Cyl Deact? 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Cyl Deact?', 'mapping': {'Y': 0, 'N': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Ordinal encoding for Var Valve Timing? 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Var Valve Timing?', 'mapping': {'N': 0, 'Y': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Var Var Valve Timing Desc
new_data['Var Valve Timing Desc'] = new_data['Var Valve Timing Desc'].astype('category') # converting type of columns to 'category'
new_data['Var Valve Timing Desc'] = new_data['Var Valve Timing Desc'].cat.codes # Assigning numerical values and storing in another column

# Ordinal encoding for Var Valve Timing? 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Var Valve Lift?', 'mapping': {'N': 0, 'Y': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Fuel Metering Sys Cd
new_data['Fuel Metering Sys Cd'] = new_data['Fuel Metering Sys Cd'].astype('category') # converting type of columns to 'category'
new_data['Fuel Metering Sys Cd'] = new_data['Fuel Metering Sys Cd'].cat.codes # Assigning numerical values and storing in another column

# Label encoding for Fuel Metering Sys Desc
new_data['Fuel Metering Sys Desc'] = new_data['Fuel Metering Sys Desc'].astype('category') # converting type of columns to 'category'
new_data['Fuel Metering Sys Desc'] = new_data['Fuel Metering Sys Desc'].cat.codes # Assigning numerical values and storing in another column

# Frequency/count encoding for Oil Viscosity
encoder = ce.CountEncoder(cols=['Oil Viscosity'])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Label encoding for Stop/Start System (Engine Management System) Code 
encoder = ce.OrdinalEncoder(mapping=[{'col': 'Stop/Start System (Engine Management System) Code', 
                                      'mapping': {'N': 0, 'Y': 1}}])
encoder.fit(new_data)
new_data = encoder.transform(new_data)

# Dealing with missing values (NaN)
new_data['Max Ethanol % - Gasoline'] = new_data['Max Ethanol % - Gasoline'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['Max Biodiesel %'] = new_data['Max Biodiesel %'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['2Dr Pass Vol'] = new_data['2Dr Pass Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['2Dr Lugg Vol'] = new_data['2Dr Lugg Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['4Dr Pass Vol'] = new_data['4Dr Pass Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['4Dr Lugg Vol'] = new_data['4Dr Lugg Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['Htchbk Pass Vol'] = new_data['Htchbk Pass Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['Htchbk Lugg Vol'] = new_data['Htchbk Lugg Vol'].replace(np.nan, 0)  # replacing NaN values with 0
new_data['GHG 1-10 rating on Ethanol (EPA Determined)'] = new_data['GHG 1-10 rating on Ethanol (EPA Determined)'].replace(np.nan, 0)  # replacing NaN values with 0
```

### <font color = green> Objective </font>
#### The main objective of this project is to predict 'Unadjusted combined fuel economy'(Comb Unadj FE - Conventional Fuel), which is a weighted average of city fuel economy and highway economy (55% and 45%, respectively).


### <font color = green> Deliverables: </font>
#### A clean dataset with all numeric and no null values, which can be used to train a Random forest model.

### <font color = green> Steps used for preparing the data </font>
- #### The columns we removed were either of the following - 
 ####  1) Did not carry any information useful for the model (example: Model year, index, or Fuel Unit Desc - Conventional Fuel)
 ####  2) They are two similar columns that were identical, so we removed one of them (example - Air Aspir Method and Air Aspiration             Method Desc)    
 ####  3)Not related to the objective or indirectly giving target information (ex - Hwy Unadj FE - Conventional Fuel)

- #### Depending upon the data type, variable type(ordinal or nominal), and the number of categories, we choose the kind of encoding to be used. We also took help from the lecture notes and a few websites when the answer was not that obvious. 


- #### If the feature had different categories with no order, we used label encoding or frequency encoding to convert them into the numeric format. 
- #### Using frequency encoding for the nominal variable is not the best option as the model will assume that a variable with 1 label is less valuable than a variable with 2 labels. But on the flip side, we are avoiding using one-hot encoding which can lead to dimensionality problems, so it's a fair trade-off.

- #### If the feature had different categories with an order or possible order, we have used ordinal encoding. The label with the most importance is given the highest number and the least important label is given the smallest number.
- #### There were few cases where genreal order is difficult so we have made few assumptions. For instance, Trans Desc here we have assumed this order 'Continuously Variable'> 'Selectable CV' > 'Automatic' > 'Automated Manual- Selectable' > 'Semi-Automatic' > 'Automated Manual' > 'Manual'.




- #### Few features could be used for our model but had too many NaN or missing values. For such, we used the replace() function and filled them with appropriate values(ex - Max Biodiesel % )

- #### We used the 'Transmission' column for feature extraction. Generally, vehicles with auto transmission have a better economy. So we tried to extract whether the vehicle has auto or manual transmission and used ordinal encoding to make a new column 'Auto_transmission'.

### <font color = green> What's next</font>

#### Out of all these features, many are going to be duplicates or irrelevant to the target feature. To figure out which ones are useful and which are not, we will have to train a Random forest model and look at the following things- 
 #### 1) Out of bag score for R^2
 #### 2) Relative feature importance 
 #### 3) Number of tree nodes, and
 #### 4) Median tree depth
